```python
data["Final Grade"]=round(0.25*data["Midterm Exam"]+0.25*data["Midterm Paper"]+0.25*data["Final Exam"]+0.25*data["Final Paper"])
data
```

```python
#print(data.iterrows())


for index, rows in data.iterrows():    
    if data["Final Grade"][index]>=60.0:
        data["Pass/Fail"][index]="P"
    else:
       data["Pass/Fail"][index]="F"
```
